## Coletando Dados de uma Tabela com Pandas

```python
from selenium import webdriver
import pandas as pd

driver = webdriver.Firefox()
```

```python
driver.get('http://localhost:8000/#/exemplo/8')
```

```python
from io import StringIO

string_io = StringIO(initial_value=driver.page_source)
pd.read_html(io=string_io, attrs={'id': 'tabela-usuarios'})[0]
```
